# Bar Chart Example

```python
#sample program for bar chart of relative abundances of 5 species of fossil from a locality
%pylab inline   #import matplotlib and numpy functions, set up for iPython

bardata=[121,59,16,20,81] 	#finds of each fossil
species=["Apatobolus micula", "Lingulella davisii", "Divobolus quadratus", "Tarphyteina taylori", "Bathmoleca addisoni"]
                #names of each fossil
    
colours=["red","green","blue","orange","purple"]
                #colours to be used in bar chart

bar([1,2,3,4,5],bardata,1,0, color=colours)
                # set up chart with 5 bars using my values and colours

xticks([1,2,3,4,5],species, rotation=45, style='italic')
                #add datalabels to x-axis
ylabel("Number of specimens found")
                #title for y axis
title("Brachiopod fossils from locality #23", weight='bold', color=(.5,.1,.1))
                #title for chart

savefig("brachbar.svg")   #save barchart to an svg file (for Inkscape)
show()          #show barchart
```
