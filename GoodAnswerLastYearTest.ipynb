```python
%pylab inline
from scipy.stats import *			# Import scipy.stats - tools to do stat calculations

infile = open("DykeTrends.csv", "r")		# Open the file 'DykeTrends.csv'
lines = infile.readlines()			# Convert the file into 'lines'
infile.close()					# Close the file

trend = []					# Create list to contain the trends
rock_type = []					# Create list to contain the rock types
position = []					# Create list to contain the position on the transect
dolerite_trend = []				# Create list to contain the trends of all doleritic dykes
dolerite_trend_position = []                    # Create list to contain the position of the doleritic dyke trend measuremens
felsite_trend = []				# Create list to contain the trends of all felsitic dykes
felsite_trend_position = []                     # Create list to contain the position of the felsitic dyke trend measurements

for line in lines:				# Loop over the lines and sort the data from the 3 columns
    data = line.split(',')			# into the appropriate lists that have been created
    trend.append(float(data[0]))
    rock_type.append(data[1])
    position.append(float(data[2]))
    if data[1] == "Dolerite/Basalt" :			# If the rock type is dolerite/basalt, store the trend and
        dolerite_trend.append(float(data[0]))		# position in these lists
        dolerite_trend_position.append(float(data[2]))
    if data[1] == "Felsite" :				# If the rock type is felsite, store the trend and position
        felsite_trend.append(float(data[0]))		# in these lists
        felsite_trend_position.append(float(data[2]))

# Produce a scatter plot showing the trends and positions of the Felsite and Dolerite/Basalt dykes
plot(dolerite_trend_position, dolerite_trend, 'rx', markersize = 5, zorder=3, label="Dolerite/Basalt")
plot(felsite_trend_position, felsite_trend, 'b^', markersize = 5, zorder=5, label="Felsite")
xlim(0, 3)											# x-limits
ylim(140, 210)										# y-limits
grid(linestyle='--', linewidth=0.5, color="black", zorder=1)				# add grid
xticks([0.0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0], [0.0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0])		# define x-ticks
yticks(range(145, 200, 5), range(145, 200, 5))						# define y-ticks
xlabel("Distance Along Transect From The West / km")					# add x-axis label
ylabel("Trend / degrees")									# add y-axis label
title("Scatter Plot Showing Trends Of  Dykes Along The Transect Line", weight="bold")	# add title
legend(title="Rock Types", loc="upper left", numpoints=1)					# define legend
show()											# show graph

# D'Agostino & Pearson Normality Test for both the positions and trends
z_position, p_position = normaltest(position)
z_trend, p_trend = normaltest(trend)

# Using a 5% significance level decided weather the positions and trends are normally distributed or not
# This test is two tailed and the data is two tailed
if p_position > 0.05:
    print "The positions are normally distributed with a probability of normal distribution of %.3f" %  p_position
else:
    print "The positions are not normally distributed with a probability of normal distribution of %.3f" %  p_position 

if p_trend > 0.05:
    print "The trends are normally distributed with a probability of normal distrution of %.3f" % p_trend
else:
    print "The trends are not normally distributed with a probability of normal distribution of %.3f" % p_trend

# Print my two hypothesise
print """
H0: The trends of the dykes do not increase towards the East of the transect	
H1: The trends of the dykes increase towards the East of the transect	
"""
# Beacuse the trends are normally distributed but the positions are not normally distributed I will use
# the Spearman's Correlation Coefficient to prove any correlation betweem the two
# This is a one-tailed test
s_all_rocktypes, p_all_rocktypestmp = spearmanr(position, trend)
s_felsite, p_felsitetmp = spearmanr(felsite_trend_position, felsite_trend) 
s_dolerite, p_doleritetmp = spearmanr(dolerite_trend_position, dolerite_trend)

# Because the spearmans test is two-tailed these values must be halved
p_all_rocktypes = p_all_rocktypestmp / 2.0
p_felsite = p_felsitetmp / 2.0
p_dolerite = p_doleritetmp / 2.0

# Print the correct hypothesis for all rock type dykes
if p_all_rocktypes > 0.05:
    print """ 
The probability of the trends increasing towards the East of the transect is %.3f
Accept H0: The trends of the dykes do not increase towards the East of the transect """ % p_all_rocktypes
else:
    print """ The probability of the trends increasing towards the East of the transect is %.3f
Accept H1: There is significant evidence that the trend of the dykes does increase towards the East of the transect """ % p_all_rocktypes

# Print the correct hypothesis for the Felsite dykes
if p_felsite > 0.05:
    print """ 
The probability of the trends of the FELSITE dykes increasing towards the East is %.3f
Accept H0: The trends of the FELSITE dykes does not increase towards the East of the transect """ % p_felsite
else:
    print """ The probability of the trends of the FELSITE dykes increasing towards the East is %.3f
Accept H1: There is significant evidence that the trends of the FELSITE dykes does increase towards the East of the transect """ % p_felsite
 
# Print the correct hypothesis for the Dolerite dykes
if p_dolerite > 0.05:
    print """ 
The probability of the trends of the DOLERITE dykes increasing towards the East is %.3f
Accept H0: The trends of the DOLERITE dykes do not increase towarsd the East of the transect 
""" % p_dolerite
else:
    print """ The probability of the trends of the DOLERITE dykes increasing towarsd the East is %.3f
Accept H1: There is significant evidence that the trends of the DOLERITE dykes does increase towards the East of the transect 
""" % p_dolerite

# Calculate and print the means of the trends
print "The mean trend of the FELSITE dykes is: %.1f" % mean(felsite_trend)
print "The mean trend of the DOLERITE dykes is: %.1f" % mean(dolerite_trend)

# The trends are normally distributed so Student's T-test can be used 
# The data is two-tailed and the test is two-tailed
# Print my two hypothesis
print """ 
H0: The means of the trends are not significantly different
H1: The means of the trends are significantly different """

# Calculate the t-test coefficient
t, p = ttest_ind(felsite_trend, dolerite_trend)

# Print the correct hypothesis for the means of the trend
if p > 0.05:
    print """ 
The probability of the means of the trends being the same is %.3f
Accept H0: The means of the trends are not significantly different """ % p
else:
    print """The probability of the means of the trends being the same is %.3f
Accept H1: The means of the trends are significantly different 
""" % p

# The positions are nor normally distributed so Mann-Whitney U-test must be used
# The data is two tailed but the test is only one tailed
print """H0: The means of the positions are not significantly different
H1:The means of the positions are significantly different """

# Calculate the mann-whitney coefficient
u, p = mannwhitneyu(felsite_trend_position, dolerite_trend_position)
p = p*2  # To adjust for 2-tailed test

# Print the correct hypothesis for the mean of the positions
if p > 0.05:
    print """
The probability of the means of the positions not being significantly different is %.3f
Accept H0: The means of the positions are the same """ % p
else:
    print """The probability of the means of the positions being significantly different is %.3f
Accept H1: The means of the positions are significantly different 
""" % p

# The 6 statements from the handout
A = "A. There is strong evidence supporting model A"
B = "B. There is insufficient evidence to reject model B"
C = "C. The data suggests that only one dyke type follows model A"
D = "D. Differences in the trend and position of the two dyke types generates a misleading correlation when the whole\
dataset is considered"
E = "E. The similarity of positions between the two dyke types means they do not affect any conclusions"
F = "F. The two dyke types independantly model A"

# Print the statements that agree with my data
print "The data fits the following conclusion:"
if p_all_rocktypes > 0.05 and p_felsite > 0.05 and p_dolerite > 0.05:
    print B
if p_all_rocktypes < 0.05 and p_felsite < 0.05 and p_dolerite < 0.05:
    print A
if p_felsite > 0.05 and  p_dolerite < 0.05:
    print C
if p_dolerite > 0.05 and p_felsite < 0.05:
    print C
if p_all_rocktypes < 0.05 and p_felsite > 0.05 and p_dolerite > 0.05:
    print D
if p_felsite < 0.05 and p_dolerite > 0.05:
    print E
if p_dolerite < 0.05 and p_felsite > 0.05:
    print E
if p_dolerite <0.05 and p_felsite < 0.05:
    print A

# I think that the correct statement is D. This is because when assesing all of the data together it looks as though
# the trends do increase towards the East. But this isnt possible when individually the trend doesnt increase for 
# either of the rock types individually.
# B is also correct as there is not enough evidence to reject model B
```
